## Kod Açıklaması

### `x = np.arange(0, 10)`

Bu satırda NumPy kütüphanesinin `arange` fonksiyonu kullanılarak 0'dan 10'a kadar olan sayıları içeren bir dizi oluşturulur.

### `fig, ax = plt.subplots(2, 2, figsize =(12, 6))`

Bu satırda, 2x2 boyutunda bir figür ve alt grafik (subplot) oluşturulur. Figürün boyutu 12x6 olarak belirlenir.

### `plt.suptitle("LineWidth & Marker & MarkerSize & Color", fontsize = 16, y = 0.95)`

Bu satırda figüre bir üst başlık (suptitle) eklenir. Font büyüklüğü 16 olarak, ve y konumu 0.95 olarak belirlenir.

### `ax[0,0].plot(x, "r", marker = "+", markersize="5")` ... `ax[1,1].plot(x-4, "r", marker = "1", ms="5");`

Bu satırlar 2x2 boyutundaki alt grafiklerin her birine bir çizgi grafiği çizer. Çizgilerin rengi, marker tipi, marker boyutu ve çizgi kalınlığı (linewidth) farklı şekillerde belirlenir.

## Output Açıklaması

Bu kod bloğu çalıştırıldığında, 2x2 bir ızgara şeklinde 4 farklı alt grafik oluşturulur. Her biri farklı bir çizgi grafiği içerir. Çizgi grafikleri farklı renklerde, farklı marker tipleriyle ve farklı çizgi kalınlıklarıyla oluşturulur.

Bu kod bloğu, çizgi grafiklerinin çeşitli görsel özelliklerini nasıl değiştirebileceğinizi gösterir.

```python
x = np.arange(0, 10)

fig, ax = plt.subplots(2, 2, figsize =(12, 6))

plt.suptitle("LineWidth & Marker & MarkerSize & Color", fontsize = 20, y = 1)

ax[0,0].plot(x, "r", marker = "+", markersize="5")
ax[0,0].set_title("Grafik-1", fontsize=18)
ax[0,0].set_xlabel("X", fontsize=18)
ax[0,0].set_ylabel("Y")

ax[0,1].plot(x-2, "y", marker = "o", markersize="5")
ax[0,1].set_title("Grafik-2", fontsize=18)
ax[0,1].set_xlabel("X", fontsize=18)
ax[0,1].set_ylabel("Y", fontsize=18)

ax[1,0].plot(x-3, "black", marker = "s", markersize="5", lw=0)
ax[1,0].set_title("Grafik-3", fontsize=18)
ax[1,0].set_xlabel("X", fontsize=18)
ax[1,0].set_ylabel("Y", fontsize=18)

ax[1,1].plot(x-4, "r", marker = "1", ms="5")
ax[1,1].set_title("Grafik-4", fontsize=18)
ax[1,1].set_xlabel("X", fontsize=18)
ax[1,1].set_ylabel("Y", fontsize=18)

plt.tight_layout();
```

```python
x = np.arange(-50, 50)
y1 = np.power(x, 2)
y2 = np.power(x, 3)
y3 = np.sin(x)
y4 = np.cos(x)
y5 = np.tan(x)
y6 = np.tanh(x)
y7 = np.sinh(x)
y8 = np.cosh(x)
y9 = np.exp(x)



fig1, ax1 = plt.subplots(nrows = 3, ncols = 3, figsize= (20, 20))

ax1[0, 0].plot(x, y1, "tab:blue") # set the color of line chart
ax1[0, 0].set_title("Square Function") # setting title of subplot
ax1[0, 0].set_xlabel("X", fontsize = 18)
ax1[0, 0].set_ylabel("Y", fontsize = 18)

ax1[0, 1].plot(x, y2, "tab:orange") # set the color of line chart
ax1[0, 1].set_title("Cubic Function") # setting title of subplot
ax1[0, 1].set_xlabel("X", fontsize = 18)
ax1[0, 1].set_ylabel("Y", fontsize = 18)

ax1[0, 2].plot(x, y3, "tab:green") # set the color of line chart
ax1[0, 2].set_title("Sine Function") # setting title of subplot
ax1[0, 2].set_xlabel("X", fontsize = 18)
ax1[0, 2].set_ylabel("Y", fontsize = 18)

ax1[1, 0].plot(x, y4, "r-") # set the color of line chart
ax1[1, 0].set_title("Cosine Function") # setting title of subplot
ax1[1, 0].set_xlabel("X", fontsize = 18)
ax1[1, 0].set_ylabel("Y", fontsize = 18)

ax1[1, 1].plot(x, y5, "b-") # set the color of line chart
ax1[1, 1].set_title("Tangent Function") # setting title of subplot
ax1[1, 1].set_xlabel("X", fontsize = 18)
ax1[1, 1].set_ylabel("Y", fontsize = 18)

ax1[1, 2].plot(x, y6, "g-") # set the color of line chart
ax1[1, 2].set_title("Hyperbolic Tangent Function") # setting title of subplot
ax1[1, 2].set_xlabel("X", fontsize = 18)
ax1[1, 2].set_ylabel("Y", fontsize = 18)

ax1[2, 0].plot(x, y7, "m-") # set the color of line chart
ax1[2, 0].set_title("Hyperbolic Sine Function") # setting title of subplot
ax1[2, 0].set_xlabel("X", fontsize = 18)
ax1[2, 0].set_ylabel("Y", fontsize = 18)

ax1[2, 1].plot(x, y8, "y*") # set the color of line chart
ax1[2, 1].set_title("Hyperbolic Cosine Function") # setting title of subplot
ax1[2, 1].set_xlabel("X", fontsize = 18)
ax1[2, 1].set_ylabel("Y", fontsize = 18)

ax1[2, 2].plot(x, y9, "kv") # set the color of line chart
ax1[2, 2].set_title("Exp Function") # setting title of subplot
ax1[2, 2].set_xlabel("X", fontsize = 18)
ax1[2, 2].set_ylabel("Y", fontsize = 18)

plt.show()
```

## Pie Chart

## Kod Açıklaması

### `x = np.array([100, 100, 100, 100])` 

Bu satırda, her bir meyve türü için 100 birimlik bir dizi oluşturulur.

### `mylabels = ["Apples", "Bananas", "Cherries", "Melon"]`

Bu satırda, pasta dilimlerini etiketlemek için bir liste oluşturulur. Önceki örnekte "Dates" vardı, bu sefer "Melon" eklenmiş.

### `plt.figure(figsize=(10, 8))`

Bu satırda, çizilecek figürün boyutları 10x8 olarak belirlenir.

### `plt.pie(x, labels = mylabels, labeldistance = 0.75, autopct = "%.2f")`

Bu satırda, pasta grafiği çizilir. `labels` parametresi ile dizi elemanlarını etiketleriz, `labeldistance` ile etiketlerin merkeze olan uzaklığı ayarlanır, `autopct` ile yüzde olarak ne kadar alan kapladığı gösterilir.

## Output Açıklaması

Bu kod bloğu çalıştırıldığında, dört farklı meyve türü ("Apples", "Bananas", "Cherries", "Melon") için eşit boyutlu pasta dilimleri oluşturulur. Her bir dilimin yüzde olarak ne kadar alan kapladığı etiketlenir. 

Bu kod bloğu, meyve türlerinin birbiri arasındaki oransal dağılımı görselleştirmek için kullanılır. Bu sefer meyve türleri arasına "Melon" eklenmiştir.

```python
x = np.array([100, 100, 100, 100])
mylabels = ["Apples", "Bananas", "Cherries", "Melon"]

plt.figure(figsize=(10, 8))
plt.pie(x, labels = mylabels, labeldistance = 0.50, autopct = "%.2f");
```

```python
x = np.array([100, 100, 100, 100])
mylabels = ["Apples", "Bananas", "Cherries", "Melon"]

myexplode =[0.1, 0, 0, 0]

plt.figure(figsize=(10, 8))
plt.pie(x, labels = mylabels, labeldistance = 0.75, autopct = "%.2f");
```

```python
z = np.array([16, 21, 45, 19])
mylabels = ["Apples", "Bananas", "Cherries", "Melon"]

myexplode =[0.1, 0, 0, 0]

plt.figure(figsize=(10, 8))
plt.pie(z, labels = mylabels, explode= myexplode, 
        labeldistance = 0.74, 
        autopct = "%.4f", 
        startangle = 90);
```

## Kod Açıklaması

### `plt.figure(figsize=(10,8))`

Bu satırda, çizilecek figürün boyutları 10x8 olarak belirlenir.

### `color_list= ["gold", "yellowgreen", "lightcoral", "lightskyblue"]`

Bu satırda, her bir pasta dilimi için kullanılacak renkler bir liste içinde tanımlanır.

### `explode_list = [0.5, 0.5, 0.5, 0.5]`

Bu satırda, her bir pasta diliminin merkezden ne kadar uzaklaştırılacağını belirleyen bir liste oluşturulur.

### `y= np.array([25, 35, 25, 15])`

Bu satırda, her bir meyve türü için farklı sayıda birimler belirlenir.

### `mylabels = ["Apples", "Bananas", "Cherries", "Melon"]`

Bu satırda, pasta dilimlerini etiketlemek için bir liste oluşturulur.

### `plt.pie(...)`

Bu satırda, pasta grafiği çizilir. `labels`, `labeldistance`, `autopct`, `startangle`, `shadow`, `pctdistance`, `colors`, ve `explode` parametreleri ile çeşitli görsel ayarlar yapılır.

### `plt.legend(loc = (0))`

Bu satırda, grafiğe bir gösterge (legend) eklenir.

### `plt.show()`

Bu satırda, çizilen grafik ekrana getirilir.

## Output Açıklaması

Bu kod bloğu çalıştırıldığında, dört farklı meyve türü ("Apples", "Bananas", "Cherries", "Melon") için pasta dilimleri oluşturulur. Her bir dilimin yüzde olarak ne kadar alan kapladığı, belirlenen renklerle ve merkezden belirli bir miktar uzaklaştırılarak etiketlenir. Ayrıca, bir gölge efekti de eklenir.

Bu kod bloğu, meyve türlerinin birbiri arasındaki oransal dağılımı görselleştirmek için kullanılır, ancak daha fazla görsel ayar ve detay eklenmiştir.

```python
plt.figure(figsize=(10, 8))

color_list= ["gold", "yellowgreen", "lightcoral", "lightskyblue"]
explode_list = [0.5, 0.5, 0.5, 0.5]

y= np.array([25, 35, 25, 15])

mylabels = ["Apples", "Bananas", "Cherries", "Melon"]

plt.pie(y, labels = mylabels, 
        labeldistance = 0.7, 
        autopct = "%.1f", 
        startangle = 90, 
        shadow = True, 
        pctdistance = 0.6,
       colors = color_list,
       explode = explode_list)

plt.legend(loc = (0))
plt.show()
```
