# Çoklu data ve legend verme

```python
age = np.arange(25, 44, 2.5)
salary_1= np.arange(40000, 76000, 5000)
salary_2= [45372, 46852, 69401, 56790, 73000, 80000, 83545, 95653] 
```

```python
plt.plot(age, salary_1, "b:", label = "Salary_1")
plt.plot(age, salary_2, "r-", label = "Salary_2")

plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("SALARY BY AGE")

plt.legend()

plt.show()
```

```python
plt.plot(age, salary_1, "g:", label = "Salary_1")
plt.plot(age, salary_2, "m-", label = "Salary_2")

plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("SALARY BY AGE")

plt.legend(loc='lower right')

plt.show()
```

```python
plt.plot(age, salary_1, "g:", label = "Salary_1")
plt.plot(age, salary_2, "m-", label = "Salary_2")

plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("SALARY BY AGE")

plt.legend(bbox_to_anchor=(0.6, 1))

plt.show()
```

```python
plt.plot(age, salary_1, "g:", label = "Salary_1")
plt.plot(age, salary_2, "m-", label = "Salary_2")

plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("SALARY BY AGE")

plt.legend(loc='upper right')

plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(age, salary_1, "b-", linewidth=6, label = "Salary 1")
plt.plot(age, salary_2, "r-.", lw=2, label = "Salary 2")

plt.legend()
plt.show()
```

# Figure ve Axes Renklendirmesi

# Function Method and Object Oriented Methods

## Functional Method

```python
plt.style.use("seaborn-darkgrid")

plt.plot(age, salary_1)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Salary by Age")

plt.show()
```

## Object Oriented Method

```python
fig, ax = plt.subplots()

ax.plot(age, salary_1, "b")
ax.set_xlabel("age")
ax.set_ylabel("Salary")
ax.set_title("Salary by Age");
```

## Functional Method (çoklu grafik çizimi)

```python
plt.subplot(2, 1, 1) # 2 satırlık 1 sütunluk figür ün 1. grafiği
plt.plot(age, salary_1, "r")

plt.subplot(2, 1, 2) # 2 satırlık 1 sütunluk figür ün 2. grafiği
plt.plot(age, salary_2, "b");
```

## Object Oriented Method (Çoklu Grafik Çizme)

```python
fig, ax = plt.subplots(2, 1)

ax[0].plot(age, salary_1, "r")
ax[1].plot(age, salary_2, "b");
```

```python
fig, ax = plt.subplots(2, 1)

ax[0].plot(age, salary_1, "r")
ax[1].plot(age, salary_2, "b")

plt.tight_layout(); # grafiklerin birbirine girmesini engeller
```

## Functional Method (çoklu grafik çizimi)

```python
plt.subplot(2,3,1)
plt.plot(age, salary_1, "r")

plt.subplot(2,3,2)
plt.plot(age, salary_1, "r")

plt.subplot(2,3,3)
plt.plot(age, salary_1, "r")

plt.subplot(2,3,4)
plt.plot(age, salary_1, "b")

plt.subplot(2,3,5)
plt.plot(age, salary_1, "b")

plt.subplot(2,3,6)
plt.plot(age, salary_1, "b")

plt.show()
```

```python
plt.subplot(2,3,1)
plt.plot(age, salary_1, "r")

plt.subplot(2,3,2)
plt.plot(age, salary_1, "r")

plt.subplot(2,3,3)
plt.plot(age, salary_1, "r")

plt.subplot(2,3,4)
plt.plot(age, salary_1, "b")

plt.subplot(2,3,5)
plt.plot(age, salary_1, "b")

plt.subplot(2,3,6)
plt.plot(age, salary_1, "b")

plt.tight_layout(); ## Grafiklerin birbirinin içine girmesini engeller
plt.show()
```

## Object Oriented Method (Çoklu Grafik Çizme)

```python
fig, ax = plt.subplots(2, 3)

ax[0][0].plot(age, salary_1, "r")
ax[0][1].plot(age, salary_1, "r")
ax[0][2].plot(age, salary_1, "r")

ax[1][0].plot(age, salary_1)
ax[1][1].plot(age, salary_1)
ax[1][2].plot(age, salary_1)

plt.show()
```

```python
fig, ax = plt.subplots(2, 3)

ax[0][0].plot(age, salary_1, "r")
ax[0][1].plot(age, salary_1, "r")
ax[0][2].plot(age, salary_1, "r")

ax[1][0].plot(age, salary_1)
ax[1][1].plot(age, salary_1)
ax[1][2].plot(age, salary_1)

plt.tight_layout()
plt.show()
```
